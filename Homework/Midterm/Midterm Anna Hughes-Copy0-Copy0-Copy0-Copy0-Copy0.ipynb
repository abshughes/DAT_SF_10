```python
# Function for extracting the title of a person

def title_get(df) :

    # getting the piece of string which indicates title
    for i in df.index: 
        df.ix[i, 'Title']=df.ix[i, 'Name'].partition('.')[0]
        df.ix[i, 'Title']=df.ix[i, 'Title'].partition(',')[2]
        
    # list of possible abbreviations for titles
    title_names=list(pd.unique(df['Title']).ravel())
    
    # return the title abbreviations
    return title_names
```

```python
# Function for transforming all kinds of titles into four basic types: Mrs, Mr, Miss and Master. 
# This function also assigns the marital status (0 for unmarried and 1 for married)

def substrings_in_string(big_string, substrings):
    for substring in substrings:
        if string.find(big_string, substring) != -1:
            return substring
    print big_string
    return np.nan

# Creating a new column 'Title'
def title_transform(df, title_names):
    df['Title']=df['Name'].map(lambda x: substrings_in_string(x, title_names))
    a=df['Title']
    return df

# Creating a new column 'Martial_status'
def martial_status(df):
    title_names_Mr=[' Don', ' Major', ' Capt', ' Jonkheer', ' Rev', ' Col', ' Mr', ' Sir', 'Mr']
    title_names_Mrs=[' the Countess', ' Mme', ' Mrs',' Lady', 'Mrs']
    title_names_Miss=[' Mlle', ' Ms',' Miss', 'Miss']
    title_names_Master=[' Master', 'Master']
    for i in df.index: 
        if df.ix[i, 'Title'] in title_names_Mr:
            df.ix[i, 'Title']='Mr'
            df.ix[i, 'Title_mar']=1
        elif df.ix[i, 'Title'] in title_names_Mrs:
            df.ix[i, 'Title']='Mrs'
            df.ix[i, 'Title_mar']=1
        elif df.ix[i, 'Title'] in title_names_Miss:
            df.ix[i, 'Title']='Miss'
            df.ix[i, 'Title_mar']=0
        elif df.ix[i, 'Title'] in title_names_Master:
            df.ix[i, 'Title'] = 'Master'
            df.ix[i, 'Title_mar']=0
        elif df.ix[i, 'Title']in [' Dr', 'Dr']:
            if df.ix[i, 'Sex']=='male':
                df.ix[i, 'Title']='Mr'
                df.ix[i, 'Title_mar']=1
            else:
                df.ix[i, 'Title']='Mrs'
                df.ix[i, 'Title_mar']=1
        return df
```

I just used a cross validaton function

```python
from sklearn.cross_validation import KFold

def cross_validate(X, y, classifier, k_fold) :

    # derive a set of (random) training and testing indices
    k_fold_indices = KFold(len(X), n_folds=k_fold,
                           shuffle=True, random_state=0)

    k_score_total = 0
    # for each training and testing slices run the classifier, and score the results
    for train_slice, test_slice in k_fold_indices :

        model = classifier(X[ train_slice  ],
                         y[ train_slice  ])
        k_score = model.score(X[ test_slice ],
                              y[ test_slice ])

        k_score_total += k_score

    # return the average accuracy
    return k_score_total/k_fold
```

```python
model_lr = LogisticRegression(C=1).fit

cross_validate(features, target, model_lr, 40)

```
