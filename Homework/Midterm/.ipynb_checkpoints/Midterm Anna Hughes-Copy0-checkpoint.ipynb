###### Importing data

There is no tickets which cost 0. So I replace 0 with NaN

```python
import pandas as pd
import numpy as np
import string
import statsmodels.api as sm
df=pd.read_csv('train.csv')
```

###### Extracting Titles from Names

```python
for i in df.index: 
    df.ix[i, 'Title']=df.ix[i, 'Name'].partition('.')[0]
    df.ix[i, 'Title']=df.ix[i, 'Title'].partition(',')[2]
Title_names=list(pd.unique(df['Title'].values.ravel()))
print Title_names
```

Replacing titles with codes
0 for male, 1 for female (Title_sex)
0 for young, 1 for mature (Title_age)
There are 4 titles left Mr, Mrs, Miss and Master

```python
for i in df.index: 
    if df.ix[i, 'Title'] in [' Don', ' Major', ' Capt', ' Jonkheer', ' Rev', ' Col', ' Mr', ' Sir', 'Mr']:
        df.ix[i, 'Title']='Mr'
        df.ix[i, 'Title_sex']=0
        df.ix[i, 'Title_age']=1
    elif df.ix[i, 'Title'] in [' the Countess', ' Mme', ' Mrs',' Lady', 'Mrs']:
        df.ix[i, 'Title']='Mrs'
        df.ix[i, 'Title_sex']=1
        df.ix[i, 'Title_age']=1
    elif df.ix[i, 'Title'] in [' Mlle', ' Ms',' Miss', 'Miss']:
        df.ix[i, 'Title']='Miss'
        df.ix[i, 'Title_sex']=1
        df.ix[i, 'Title_age']=0
    elif df.ix[i, 'Title'] in [' Master', 'Master']:
        df.ix[i, 'Title'] = 'Master'
        df.ix[i, 'Title_sex']=0
        df.ix[i, 'Title_age']=0
    elif df.ix[i, 'Title']in [' Dr', 'Dr']:
        if df.ix[i, 'Sex']=='male':
            df.ix[i, 'Title']='Mr'
            df.ix[i, 'Title_sex']=0
            df.ix[i, 'Title_age']=1
        else:
            df.ix[i, 'Title']='Mrs'
            df.ix[i, 'Title_sex']=1
            df.ix[i, 'Title_age']=1
            
Title_names=list(pd.unique(df['Title'].values.ravel()))
print Title_names
```

Here I am going to replace Sex with a 0/1 value (0 male, 1 female)

```python
df['Sex_01']=df['Sex'].map({'female':1, 'male':0}).astype(int)
```

Checking whether the initial Sex values corresponds to the one derived from Title

```python
for i in df.index: 
    if df.ix[i,'Sex_01']<>df.ix[i,'Title_sex']:
        print df.ix[i,'Name']
for i in df.index: 
    if df.ix[i,'Title_age']==0:
        print df.ix[i,'Name']
               
# nothing is printed because all data is consistent
```

I will use 3 dimensions to calculate the average age for. 
1. Sex (equal to title_sex)
2. Title_age
3. Cabin class

```python
median_ages = np.zeros((2,2,3))
for i in range(0, 2):
    for j in range(0, 2):
        for k in range (0,3):
            median_ages[i,j,k] = df[(df['Title_sex'] == i) &(df['Title_age'] == j) &(df['Pclass'] == k+1)]['Age'].dropna().median()
median_ages[:,0,:]
df['Age'].isnull()
```

```python
for i in range(0, 2):
    for j in range(0, 2):
        for k in range (0,3):
            df.loc[ (df.Age.isnull()) & (df['Title_sex'] == i)&(df['Title_age'] == j) & (df.Pclass == k+1),\
                'Age'] = median_ages[i,j,k]
df['Age'].isnull()
```

###### Cabins

Hypothesis: the probvability of survival is affected by the location of the cabin. And location of the cabin is defined by its letter(A-F) and number (two-digit after letter)

```python
cabin_names=list(pd.unique(df['Cabin'].values.ravel()))
#print cabin_names.sort()

df['Cabin']=df['Cabin'].astype(str)
# df.info()
# print cabin_names, len(cabin_names)

for i in df.index: 
    cabin_code=df.ix[i, 'Cabin']
    if cabin_code<>'nan':
        df.ix[i, 'Cabin_letter']=cabin_code[0]
        df.ix[i, 'Cabin_number']=cabin_code[1:4]
    else:
        df.ix[i, 'Cabin']=np.nan
        df.ix[i, 'Cabin_letter']=np.nan
        df.ix[i, 'Cabin_number']=np.nan
        # print df.head()
        
df_cabins=df.dropna(subset=['Cabin'], how='any')
#df_cabins
```
