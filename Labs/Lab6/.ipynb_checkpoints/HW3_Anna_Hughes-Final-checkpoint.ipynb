### Importing the csv_file

```python
import pandas as pd
import numpy as np
data_log=pd.read_csv("student_logins.csv")
data_log[:5]
```

In this homework I'm going to run regression for multiple variables. 

### Processing data to get nessesary variables

###### Converting session_start_time and session_end_time to datetime type

```python
from datetime import datetime
data_log['startdatetime'] = data_log['session_start_time'].map(lambda x: datetime.strptime(x, "%m/%d/%y %H:%M"))
data_log['startdatetime'].head(5)  
```

```python
data_log['enddatetime'] = data_log['session_end_time'].map(lambda x: datetime.strptime(x, "%m/%d/%y %H:%M"))
data_log['enddatetime'].head(5)   
```

```python
data_log['accounttime'] = data_log['student_account_created'].map(lambda x: datetime.strptime(x, "%m/%d/%y %H:%M"))
data_log['accounttime'].head(5)  
```

###### Creating the variable 'duration'

```python
data_log['duration']=data_log['enddatetime']-data_log['startdatetime']
data_log['duration'].head(5)
```

###### Creating the variable 'time_exists' (how long the accound exists)

```python
data_log['time_exists']=data_log['startdatetime']-data_log['accounttime']
data_log['time_exists'].head(5)
```

###### Conversions for 'duration' and 'time_exists'

```python
data_log['duration']= data_log['duration'].astype("timedelta64[m]")
#data_log['duration']=data_log['duration']/np.timedelta64(1,'m')
#data_log['duration']=np.int8(data_log['duration'])
data_log['duration'].head(5)
```

```python
data_log['time_exists']= data_log['time_exists'].astype("timedelta64[m]")
data_log['time_exists'].head(5)
```

###### Creating a variable "avg_task_dur" which is average task duration

```python
data_log['avg_task_dur'] = data_log['duration']/data_log['problems_completed']
data_log['avg_task_dur'].head(5)
```

###### Regression: student_id VS time_exists

```python
#%matplotlib inline
from matplotlib import pyplot as plt
from pylab import *
plt.title('student_id VS time_exists')
plt.xlabel('student_id')
plt.ylabel('time_exists')
plt.plot(data_log['student_id'],data_log['time_exists'], 'bo')

X = sm.add_constant(data_log['student_id'], prepend=True)
results = sm.OLS(data_log['time_exists'], X).fit()
intercept, slope = results.params

r2 = results.rsquared
xl = np.array([min(data_log['student_id']), max(data_log['student_id'])])
yl = intercept + slope * xl
plt.plot(xl, yl, 'r-')
plt.show
results.summary()
```

```python
# importing packages
import statsmodels.api as sm
# setting inputs and outputs
x=data_log[['class_id_avg']]
#x = data_log[['class_id_avg','student_previous_logins_total','problems_completed', 'student_previous_logins_total','student_previous_class_logins']].values
y = data_log['avg_task_dur'].values

X = sm.add_constant(x, prepend=True)
results = sm.OLS(y, X).fit()
results.summary()
```
