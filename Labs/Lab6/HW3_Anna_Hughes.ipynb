## Build a program that can predict the duration time of an online student

### 1. Importing the csv_file

```python
import pandas as pd
import numpy as np
data_log=pd.read_csv("student_logins.csv")
data_log[:5]
```

### 2. Processing data to get nessesary variables

###### Converting session_start_time and session_end_time to datetime type

```python
from datetime import datetime
data_log['startdatetime'] = data_log['session_start_time'].map(lambda x: datetime.strptime(x, "%m/%d/%y %H:%M"))
data_log['startdatetime'].head(5)  
```

```python
data_log['enddatetime'] = data_log['session_end_time'].map(lambda x: datetime.strptime(x, "%m/%d/%y %H:%M"))
data_log['enddatetime'].head(5)   
```

```python
data_log['accounttime'] = data_log['student_account_created'].map(lambda x: datetime.strptime(x, "%m/%d/%y %H:%M"))
data_log['accounttime'].head(5)  
```

###### Creating the variable 'duration'

```python
data_log['duration']=data_log['enddatetime']-data_log['startdatetime']
data_log['duration'].head(5)
```

###### Creating the variable 'time_exists' (how long the accound exists)

```python
data_log['time_exists']=data_log['startdatetime']-data_log['accounttime']
data_log['time_exists'].head(5)
```

###### Conversions for 'duration' and 'time_exists'

```python
data_log['duration']= data_log['duration'].astype("timedelta64[m]")
#data_log['duration']=data_log['duration']/np.timedelta64(1,'m')
#data_log['duration']=np.int8(data_log['duration'])
data_log['duration'].head(5)
```

```python
data_log['time_exists']= data_log['time_exists'].astype("timedelta64[m]")
data_log['time_exists'].head(5)
```

### 4. Taking into account categorical variables

###### Getting unique values

```python
class_names=list(pd.unique(data_log['class_id'].values.ravel()))
print 'List of unique class_id is = %s' % class_names
student_names=list(pd.unique(data_log['student_id'].values.ravel()))
print 'List of unique student_id is = %s' % student_names
```

```python
data_log_class_id=pd.get_dummies(data_log['class_id'])
data_log=data_log.join(data_log_class_id)
data_log_student_id=pd.get_dummies(data_log['student_id'])
data_log=data_log.join(data_log_student_id)
data_log.head()
```

### 5. Running regression on several variables

```python
# importing packages
import statsmodels.api as sm
# setting inputs and outputs
x = data_log[['problems_completed','a', 'c', 'e', 'g','m']].values
y = data_log['duration'].values

X = sm.add_constant(x, prepend=True)
results = sm.OLS(y, X).fit()
results.summary()
```
